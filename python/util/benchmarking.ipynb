```python
Lx = 1000000
Lk = 10000

x = np.random.randn(Lx)
k = np.exp(-square((arange(Lk)-Lk/2.0)/(2*Lk/10.0)))

figure(figsize=(20,5))
plot(x);
figure(figsize=(10,5))
plot(k);
```

```python
%%timeit -n 10
y1 = np.convolve(x, k, 'same')
```

```python
figure(figsize=(20,5))
plot(y1);
```

```python
%%timeit -n 10
y2 = parallelConvolve(x, k, 300)
```

```python
figure(figsize=(20,5))
plot(y2);
```
